```python
from fastai import *
from fastai.vision import *
```

```python
#moved a few dirs to train (from data)
path = Path('/home/jupyter/mushrooms/train')
```

```python
np.random.seed(42)
data = ImageDataBunch.from_folder(path, train=".", valid_pct=0.2,
        ds_tfms=get_transforms(), size=224, num_workers=4).normalize(imagenet_stats)
```

```python
learn = create_cnn(data, models.resnet34, metrics=error_rate)
```

```python
learn.load('3_resnet34_defaults')
```

```python
interpr = ClassificationInterpretation.from_learner(learn)
```

```python
interpr.plot_confusion_matrix()
```

```python
interpr.plot_top_losses(9)
```
