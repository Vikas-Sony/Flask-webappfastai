```python
path = Path('/home/jupyter/mushrooms/')
```

```python
keywords = 'champignon,oesterzwam,gewoon eekhoorntjesbrood'
classes = keywords.split(',')
classes
```

```python
img = open_image(path/'data'/'oesterzwam'/'9. oesterzwammen.jpg')
img = open_image(path/'data'/'oesterzwam'/'8. oesterzwam-800x419.jpg')
img = open_image(path/'data'/'oesterzwam'/'50. gewone_oesterzwam1.jpg')
img = open_image(path/'data'/'gewoon eekhoorntjesbrood'/'6. gewoon%20eekhoorntjesbrood.jpg')
img
```

```python
data2 = ImageDataBunch.single_from_classes(path, classes, tfms=get_transforms(), size=224).normalize(imagenet_stats)
learn = create_cnn(data2, models.resnet34)
learn.load('3_resnet34_defaults')
```

```python
pred_class,pred_idx,outputs = learn.predict(img)
pred_class
```
